```python
file_path = 'BET File.xlsx'
xls = pd.ExcelFile(file_path)
```

```python
# Specify the sheet names
sheet_1 = 'Raw Data'
sheet_2 = 'Mapping'
```

```python
df1 = pd.read_excel(xls, sheet_1)
df2 = pd.read_excel(xls, sheet_2)
```

```python
# Extract unique values from df2
country_list = df2['Market'].unique()
bu_list = df2['Types of BU'].dropna().unique()
# Να δοκιμάσω να κάνω κάτι παρόμοιο και για το Region (Υπάρχει το value NA που σημαίνει North America και η python νομίζει ότι είναι na=not available)
```

```python
# Define a function to find matching country names using regular expressions
def find_country(key_value):
    # Combine country names into a regular expression pattern
    country_pattern = '|'.join([re.escape(country) for country in country_list])

    # Search for the pattern in the 'Key Value' column
    match = re.search(r'\b(' + country_pattern + r')\b', key_value, flags=re.IGNORECASE)

    if match:
        return match.group(0)
    else:
        return 'Country not in our list'

# Apply the function to the 'Key Value' column and update the 'Market' column
df1['Market'] = df1['Key Value'].apply(find_country)
```

```python
def find_bu(key_value):
    # Combine BU names into a regular expression pattern
    bu_pattern = '|'.join([re.escape(bu) for bu in bu_list])

    # Search for the pattern in the 'Key Value' column
    match = re.search(r'\b(' + bu_pattern + r')\b', key_value, flags=re.IGNORECASE)

    if match:
        return match.group(0)
    else:
        return 'BU not in our list'

# Apply the function to the 'Key Value' column and update the 'Types of BU' column
df1['BU'] = df1['Key Value'].apply(find_bu)
```
